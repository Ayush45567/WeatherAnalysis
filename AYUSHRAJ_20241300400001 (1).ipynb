```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/weather_data.csv")
```

```python
# Display basic information about the dataset
print("Dataset Overview:\n", df.info())
```

```python
# Display first few rows
print("\nFirst 5 Rows of the Dataset:")
print(df.head())
```

```python
# Check for missing values
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# Display basic statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Set seaborn style for better visuals
sns.set_style("whitegrid")
```

```python
# Plot temperature trends over time
plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["Temperature_C"], color="tab:red", alpha=0.7)
plt.title("Daily Temperature Trend", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Temperature (Â°C)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplot of temperature distribution by season
plt.figure(figsize=(8, 5))
sns.boxplot(x="Season", y="Temperature_C", data=df, palette="coolwarm")
plt.title("Temperature Distribution by Season")
plt.xlabel("Season")
plt.ylabel("Temperature (Â°C)")
plt.show()
```

```python
# Histogram of rainfall distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["Rainfall_mm"], bins=30, kde=True, color="blue")
plt.title("Rainfall Distribution")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Average temperature by season
plt.figure(figsize=(8, 5))
season_avg_temp = df.groupby("Season")["Temperature_C"].mean().reindex(["Winter", "Spring", "Summer", "Autumn"])
season_avg_temp.plot(kind="bar", color=["blue", "green", "red", "orange"])
plt.title("Average Temperature by Season")
plt.xlabel("Season")
plt.ylabel("Average Temperature (Â°C)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.title("Average Monthly Rainfall Trend")
plt.xlabel("Month")
plt.ylabel("Average Rainfall (mm)")
plt.xticks(range(1, 13))
plt.show()
```
